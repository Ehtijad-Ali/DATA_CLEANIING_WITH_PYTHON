### Removing Outliers using IQR method

```python
# IQR inter quantile range
dataset  = pd.read_csv('/content/drive/MyDrive/Loan-data.csv')
dataset.head(10)
```

```python
sns.boxplot(x= 'CoapplicantIncome', data=dataset)
plt.show()
```

```python
dataset.shape
```

```python
q1 = dataset['CoapplicantIncome'].quantile(0.25)
q3 = dataset['CoapplicantIncome'].quantile(0.75)
```

```python

IQR = q3-q1
```

```python
# to find maximum and minimum range
min_range = q1- (1.5*IQR)
max_range = q3+ (1.5*IQR)

min_range, max_range
```

```python
outliers = dataset[dataset['CoapplicantIncome']<=max_range]
```

```python
sns.boxplot(x= 'CoapplicantIncome', data=outliers)
plt.show()
```

```python
outliers.shape
```

### Removing Outliers by Z_score Method

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# IQR inter quantile range
dataset  = pd.read_csv('/content/drive/MyDrive/Loan-data.csv')
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.boxplot(x = 'CoapplicantIncome', data=dataset)
```

```python
# removing outliers using normal method
max_range = dataset['CoapplicantIncome'].mean() - (3*dataset['CoapplicantIncome'].std())
min_range = dataset['CoapplicantIncome'].mean() - (3*dataset['CoapplicantIncome'].std())

min_range, max_range
```

```python
new_data = dataset[dataset['CoapplicantIncome']<=max_range]
```

```python
# Calculate mean and standard deviation
mean_coapplicant_income = dataset['CoapplicantIncome'].mean()
std_coapplicant_income = dataset['CoapplicantIncome'].std()
mean_coapplicant_income, std_coapplicant_income
```

```python

# Calculate z-score
z_scores = (dataset['CoapplicantIncome'] - mean_coapplicant_income) / std_coapplicant_income
z_scores
```

```python
dataset['z_scores']= z_scores
```
